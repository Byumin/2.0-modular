# NORM 업서트 노트북

이 노트북은 `modular.db`의 `NORM` 테이블에 업서트합니다.

핵심 규칙:
- `map`은 JSON 문자열 또는 Python `dict`/`list`로 입력할 수 있습니다.
- 각 row의 `(condition.id, test.id)`가 실제 `NORMCONDITION` 테이블에 존재하는지 먼저 검증한 뒤 업서트합니다.
- `id`는 직접 입력합니다.

실행 전 준비:
1. `TEST` 데이터가 먼저 들어가 있어야 합니다.
2. `NORMCONDITION` 데이터가 먼저 들어가 있어야 합니다.


```python
# 1) 경로/스키마 설정
from pathlib import Path
import json
import sqlite3

DB_PATH = Path(r"C:\Users\user\workspace\2.0-modular\modular.db")
TABLE_NAME = 'NORM'

SCHEMA = [
    {'name': 'id', 'type': 'TEXT', 'constraints': 'NOT NULL'},
    {'name': 'name', 'type': 'TEXT', 'constraints': 'NOT NULL'},
    {'name': 'map', 'type': 'TEXT', 'constraints': 'NOT NULL'},
    {'name': 'condition.id', 'type': 'TEXT', 'constraints': 'NOT NULL'},
    {'name': 'test.id', 'type': 'TEXT', 'constraints': 'NOT NULL'},
]
PRIMARY_KEY = ['id', 'test.id', 'condition.id']

print('DB_PATH =', DB_PATH)
print('TABLE_NAME =', TABLE_NAME)
print('PRIMARY_KEY =', PRIMARY_KEY)
```

```python
# 2) NORM 테이블 구조 확인
conn = sqlite3.connect(DB_PATH)
try:
    cur = conn.execute('PRAGMA table_info("NORM")')
    print('table_info(NORM)')
    for row in cur.fetchall():
        print(row)

    print('\nforeign_key_list(NORM)')
    cur = conn.execute('PRAGMA foreign_key_list("NORM")')
    for row in cur.fetchall():
        print(row)
finally:
    conn.close()
```

## 업서트


```python
# 4) NORM 테이블 업서트
insert_columns = [col['name'] for col in SCHEMA]
schema_type_map = {col['name']: col['type'].upper() for col in SCHEMA}
required_row_keys = {'id', 'name', 'map', 'condition.id', 'test.id'}

def normalize_value(col_name: str, value):
    col_type = schema_type_map[col_name]
    if value == '':
        return None
    if col_type == 'TEXT' and isinstance(value, (dict, list)):
        return json.dumps(value, ensure_ascii=False)
    return value

def validate_map_value(raw_value):
    if raw_value in (None, ''):
        raise ValueError('map은 비어 있을 수 없습니다.')
    if isinstance(raw_value, (dict, list)):
        return raw_value
    text = str(raw_value).strip()
    if not text:
        raise ValueError('map은 비어 있을 수 없습니다.')
    try:
        json.loads(text)
    except json.JSONDecodeError as e:
        raise ValueError(f'map JSON 형식이 올바르지 않습니다: {text}') from e
    return text

normalized_rows = []

conn = sqlite3.connect(DB_PATH)
try:
    conn.execute('PRAGMA foreign_keys = ON;')
    existing_test_ids = {row[0] for row in conn.execute('SELECT "id" FROM "TEST"')}
    existing_normconditions = {
        (row[0], row[1])
        for row in conn.execute('SELECT "id", "test.id" FROM "NORMCONDITION"')
    }

    for i, row in enumerate(ROWS, start=1):
        row_keys = set(row.keys())
        missing = [k for k in required_row_keys if k not in row_keys]
        if missing:
            raise ValueError(f'{i}번째 row에 필수 컬럼 누락: {missing}')

        norm_id = str(row.get('id', '')).strip()
        norm_name = str(row.get('name', '')).strip()
        condition_id = str(row.get('condition.id', '')).strip()
        test_id = str(row.get('test.id', '')).strip()
        map_value = validate_map_value(row.get('map'))

        if not norm_id or not norm_name or not condition_id or not test_id:
            raise ValueError(f'{i}번째 row에 필수값이 비어 있습니다: {row}')

        if test_id not in existing_test_ids:
            raise KeyError(f'{i}번째 row의 test.id={test_id} 가 TEST 테이블에 없습니다.')
        if (condition_id, test_id) not in existing_normconditions:
            raise KeyError(
                f'{i}번째 row의 (condition.id={condition_id}, test.id={test_id}) 가 NORMCONDITION 테이블에 없습니다.'
            )

        materialized = {
            'id': norm_id,
            'name': norm_name,
            'map': map_value,
            'condition.id': condition_id,
            'test.id': test_id,
        }
        normalized = {k: normalize_value(k, materialized[k]) for k in insert_columns}
        normalized_rows.append(normalized)

    placeholders = ', '.join(['?'] * len(insert_columns))
    col_sql = ', '.join([f'"{c}"' for c in insert_columns])
    conflict_columns = ['id', 'test.id', 'condition.id']
    update_columns = [c for c in insert_columns if c not in conflict_columns]
    conflict_sql = ', '.join([f'"{c}"' for c in conflict_columns])
    update_sql = ', '.join([f'"{c}" = excluded."{c}"' for c in update_columns])
    insert_sql = (
        f'INSERT INTO "{TABLE_NAME}" ({col_sql}) VALUES ({placeholders}) '
        f'ON CONFLICT ({conflict_sql}) DO UPDATE SET {update_sql}'
    )
    values = [tuple(row[c] for c in insert_columns) for row in normalized_rows]

    print('실행 SQL:', insert_sql)
    print('업서트 대상 건수:', len(values))

    conn.executemany(insert_sql, values)
    conn.commit()
    print(f'업서트 완료: {len(values)}건')
finally:
    conn.close()
```
